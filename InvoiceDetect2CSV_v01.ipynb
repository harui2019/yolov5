```python
DataSet = JsonRead(AllLabelPath)
DataClasses = JsonRead(LabelClassPath)['class']
nc = len(DataClasses)

dfLabelCol = ['name'] + DataClasses
dfLabel = pd.DataFrame(columns = dfLabelCol)

if __name__ == "__main__":
    for imgIndex in DataSet.keys():
        tmpCount = [ 0 for i in range(nc) ]
        for i in DataSet[imgIndex]['class'].values():
            tmpCount[int(i['subclass_num'])] += 1
        print(tmpCount)
        
        dfLabel = pd.concat([ pd.DataFrame(
            [  [ DataSet[imgIndex]['name'] ]+tmpCount ], columns = dfLabelCol
        ), dfLabel ], ignore_index=True)
        
    dfLabel.to_csv(CSVLocationPath)
    print("'DetectSummary.csv' has saved at %s"% (OutputJsonPath+ExpFolderName))

dfLabel
```
